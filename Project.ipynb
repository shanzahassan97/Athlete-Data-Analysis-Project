```python
data['Medal_numeric'] = data['Medal'].map({'Gold': 1, 'Silver': 2, 'Bronze': 3, 'None': 0})
pysql("SELECT* FROM data")
```

```python
#ATHLETE WHO ARE OLDER WITH MORE EXPERIENCE ARE MORE LIKELY TO WIN MEDALS THAN YOUNG ATHLETES
data['Outcome'] = data['Outcome'].replace(['Lose','Win',], [0, 1])
data['Outcome'] = data['Outcome'].astype(int)
correlation_matrix = data[['Age','Outcome']].corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True, cmap='magma')
plt.show()
```

```python
#VISUALIZING MEDALS WON BY DIFFERENT AGE GROUPS
# Group by 'Age' and sum the 'Medal_numeric' to get total medals per age group
age_medal_data = data.groupby('Age')['Outcome'].sum().reset_index()

# Create a line plot
plt.figure(figsize=(10,6))
plt.plot(age_medal_data['Age'], age_medal_data['Outcome'], marker='o', color='blue')

# Adding labels and title
plt.xlabel('Age', fontsize=14)
plt.ylabel('Total Medals', fontsize=14)
plt.title('Total Medals Won by Age', fontsize=18)
plt.grid(True)
plt.xticks(np.arange(0, 100,5))
plt.yticks(np.arange(0, 7500, 500))
# Show the plot
plt.show()
```

```python
#VISUALIZING AVERAGE AGE GROUPS WON MEDAL ACROSS DIFFERENT SPORTS
# Filtering the dataset to only include rows where athletes won medals
medal_winners = data[data['Medal'].notna()]

# Grouping by sport and calculating the average age of medal winners
avg_age_medal_winners = medal_winners.groupby('Sport')['Age'].mean().reset_index()

# Sorting the sports by average age
avg_age_medal_winners = avg_age_medal_winners.sort_values('Age', ascending=False)
# Splitting the data into two halves
midpoint = len(avg_age_medal_winners) // 2
part_1 = avg_age_medal_winners[:midpoint]
part_2 = avg_age_medal_winners[midpoint:]

# Creating subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 10))

# First plot (part 1)
ax1.plot(part_1['Sport'], part_1['Age'], marker='o', color='purple')
ax1.set_title('Average Age of Medal Winners (Part 1)', fontsize=16)
ax1.set_ylabel('Average Age', fontsize=14)
ax1.set_xlabel('Sport', fontsize=14)

# Set the x-ticks and labels for part 1
ax1.set_xticks(range(len(part_1['Sport'])))
ax1.set_xticklabels(part_1['Sport'], rotation=45)

# Second plot (part 2)
ax2.plot(part_2['Sport'], part_2['Age'], marker='o', color='green')
ax2.set_title('Average Age of Medal Winners (Part 2)', fontsize=16)
ax2.set_ylabel('Average Age', fontsize=14)
ax2.set_xlabel('Sport', fontsize=14)

# Set the x-ticks and labels for part 2
ax2.set_xticks(range(len(part_2['Sport'])))
ax2.set_xticklabels(part_2['Sport'], rotation=45)
# Show the plots
plt.tight_layout()
plt.show()
```

```python
#THE PERFORMANCE BY COUNTRY AND SPLIT BY THE SEASON VARIABLE (SUMMER VS WINTER), SOME COUNTRIES MAY PERFORM BETTER IN WINTER SPORTS

# Filter data for those who have won medals (not NA)
medals_data = data[data['Medal'].notna()]

# Group by Team (Country) and Season, counting the medals
medals_by_country_season = medals_data.groupby(['NOC', 'Season']).size().reset_index(name='Medal Count')

# Pivot the data to have separate columns for Summer and Winter
medals_pivot = medals_by_country_season.pivot(index='NOC', columns='Season', values='Medal Count').fillna(0)

# Sort by total medals
medals_pivot['Total Medals'] = medals_pivot['Summer'] + medals_pivot['Winter']
medals_pivot = medals_pivot.sort_values('Total Medals', ascending=False)


# Split the dataset into 3 parts
split_size = len(medals_pivot) // 3
group_1 = medals_pivot.iloc[:split_size]
group_2 = medals_pivot.iloc[split_size:split_size*2]
group_3 = medals_pivot.iloc[split_size*2:]

# Plot for the first group
plt.figure(figsize=(15, 5))
ax1 = group_1[['Summer', 'Winter']].plot(kind='bar', stacked=False, figsize=(15, 5))
plt.title('Performance by Country: Summer vs Winter (Group 1)', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Medal Count', fontsize=14)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plot for the second group
plt.figure(figsize=(15, 5))
ax2 = group_2[['Summer', 'Winter']].plot(kind='bar', stacked=False, figsize=(15, 5))
plt.title('Performance by Country: Summer vs Winter (Group 2)', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Medal Count', fontsize=14)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plot for the third group
plt.figure(figsize=(15, 5))
ax3 = group_3[['Summer', 'Winter']].plot(kind='bar', stacked=False, figsize=(15, 5))
plt.title('Performance by Country: Summer vs Winter (Group 3)', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Medal Count', fontsize=14)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


```

```python
#WINRATE ACROSS THE COUNTRIES
# Count the total number of matches each team participated in
total_participation = data.groupby('NOC').size().reset_index(name='Total_Participation')

# Count the number of times each team won a medal
medal_wins = data[data['Medal'] != 'None'].groupby('NOC').size().reset_index(name='Medal_Wins')

# Merge the two DataFrames to get participation and wins in one table
team_stats = pd.merge(total_participation, medal_wins, on='NOC', how='left')
team_stats['Medal_Wins'] = team_stats['Medal_Wins'].fillna(0)  # Fill NaN with 0

# Calculate Win Rate (Medal Wins / Total Participation)
team_stats['Win_Rate'] = team_stats['Medal_Wins'] / team_stats['Total_Participation']
```

```python
# Sort by Win Rate for better visualization
team_stats = team_stats.sort_values('Win_Rate', ascending=False)

# Assuming 'team_stats' DataFrame from the previous code is sorted by 'Win_Rate'
# Split the data into four equal parts
split_data = np.array_split(team_stats, 4)

# Create a figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(20, 15))  # 2 rows and 2 columns for 4 parts

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Plot each part in its own subplot
for i, data_part in enumerate(split_data):
    axes[i].bar(data_part['NOC'], data_part['Win_Rate'], color='skyblue')
    axes[i].set_xlabel('Country', fontsize=12)
    axes[i].set_ylabel('Win Rate', fontsize=12)
    axes[i].set_title(f'Win Rate by Team (Part {i+1})', fontsize=14)
    axes[i].tick_params(axis='x', rotation=90)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()

```
